```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

### 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

### Line function is given as

### y = 2 + (3 * x)

### where x is an integer having values from 1 to 50

```python
y=list()
for x in range(1,51):
    y.append(2+(3*x))
x=np.arange(1,51)
plt.plot(x,y)
plt.title('Line Function y=2+(3*x)')
plt.xlabel('x values -->')
plt.ylabel('y values -->')
```

### (II) Fetch data from the created csv file to create the line plots for different values 'Open', 'High', 'Low' and 'Close' values for the given dates.

```python
'''Data is already in the dataframe called as data so here I am not creating a dataframe once again using the same dataframe'''
fig=plt.figure(figsize=(20,10))
fig.suptitle("financial data of Alphabet Inc. between October 3, 2016 to October 11, 2017")

ax1=fig.add_subplot(221)
ax1.set_title("Opening price")
ax1.plot(data['Date'],data['Open'])

ax2=fig.add_subplot(222)
ax2.set_title("Closing price")
ax2.plot(data['Date'],data['Close'])

ax3=fig.add_subplot(223)
ax3.set_title("High")
ax3.plot(data['Date'],data['High'])

ax4=fig.add_subplot(224)
ax4.set_title("Low")
ax4.plot(data['Date'],data['Low'])

fig.text(0.4,0.45, 'Alphabet Inc.',
        fontsize=45,
        color='green',
        ha='left',
        va='bottom',
        alpha=0.6
        )
```

### 4. Write a Python programming to display a bar chart of the popularity of programming Languages 
#### Sample data:
#### Programming languages: Java, Python, PHP, JavaScript, C#, C++

#### Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

### 1.) X- axis should display Languages and Y-axis their Popularity. Mark plot with proper x and y label along with title.
### 2.) X-axis should display ticks with Language names

```python
programming_languages=['Java','Python','PHP','JavaScript','C#','C++']
Popularity=[22.2,17.6,8.8,8,7.7,6.7]

fig, ax = plt.subplots()
ax.bar(programming_languages, Popularity)
ax.set_xlabel('Programming languages')
ax.set_ylabel('Popularity')
ax.set_title('Popularity of programming Languages')
#ax.grid(b=True, which='major', color='r',linestyle='-', linewidth=1)
```

### 3.) Display Grid on the figure with the following details :- type – major grid, continuous line with width as 1 and red in color

```python
fig, ax = plt.subplots()
ax.bar(programming_languages, Popularity)
ax.set_xlabel('Programming languages')
ax.set_ylabel('Popularity')
ax.set_title('Popularity of programming Languages')
ax.grid(b=True, which='major', color='r',linestyle='-', linewidth=1)
```

### 4.) Create another plot of the same data but bars should be horizontal and each bar consists of a separate color.

```python
fig, ax = plt.subplots()
colors=['orange','blue','lime','cyan','green','yellow']
ax.barh(programming_languages, Popularity,color=colors)
ax.set_xlabel('Programming languages')
ax.set_ylabel('Popularity')
ax.set_title('Popularity of programming Languages')
ax.grid(b=True, which='major', color='r',linestyle='-', linewidth=1)
```

### 5. Draw a figure consisting of two circles of diameter 3 units and 4 units respectively using matplotlib. First circle will be filled with yellow color and the second circle with slanting lines. Highlight the center of both with a marker and also annotate with center coordinates having (x,y) values.

```python
fig, ax=plt.subplots()
ax.set_aspect(aspect='equal')
plt.xlim(0.0,15.0)
plt.ylim(0.0,15.0)
#for p in [patches.Circle((4,4),3,color='yellow'),patches.Circle((10,10.5),4,fill=False,hatch="//")]:
ax.add_patch(patches.Circle((4,4),3,color='yellow'))
ax.annotate('(4,4)',xy=[4,4],
            xytext=(4,6),
           arrowprops=dict(color='g',shrink=0.18))
ax.plot([4],[4],'ro')

ax.add_patch(patches.Circle((10,10.5),4,fill=False,hatch="//"))
ax.annotate('(10,10.5)',xy=[10,10.5],
            xytext=(10,12.5),
           arrowprops=dict(color='g',shrink=0.19))
ax.plot([10],[10.5],'ro')
```
