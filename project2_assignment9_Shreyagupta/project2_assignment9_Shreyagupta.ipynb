This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. # Task 6 – Pie Chart

### A) Create a Pie chart to display the distribution of developers based on their Formal Education
#### Display their percentages on the Pie-Chart
#### The respective Formal Education should be displayed outside the circle
#### The title of the graph should be ‘Classification of Developers by their Formal Education’

```python
Formaleducation=pd.DataFrame()
Formaleducation=df_survey['FormalEducation'].value_counts()
Formaleducation
```

```python
#plt.axis('equal')
fig=plt.figure(figsize=(12,12))
labels = Formaleducation.index
sizes = Formaleducation.values
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True, startangle=150)
plt.legend(labels, loc="best")
plt.title('Classification of Developers by their Formal Education')
```

### B) Create the same Piechart as in Section A) with the following changes
#### Explode the three segments with lowest values
#### Pass in a custom set of colors for different segments

```python
fig=plt.figure(figsize=(12,12))
labels = Formaleducation.index
sizes = Formaleducation.values
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','red','orange','pink','brown','green']
explode = (0, 0, 0, 0, 0, 0, 0.2, 0.5, 0.9)  
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(labels, loc="best")
```
